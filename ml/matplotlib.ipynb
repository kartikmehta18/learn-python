sine cosine wave

```python
x= np.arange(0,5* np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)

plt.plot(x,y_sin,"b--")
plt.title("sine")

plt.subplot(2,1,2)
plt.plot(x,y_cos,"g")
plt.title('cos wave')



```
